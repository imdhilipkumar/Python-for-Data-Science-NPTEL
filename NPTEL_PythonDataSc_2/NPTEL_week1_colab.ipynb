the precedence of operators is and > or > not
and
Returns True if both conditions are true. For example, (x > 3) and (y < 15) returns True if both x > 3 and y < 15 are true.
or
Returns True if at least one condition is true. For example, (x > 10) or (y < 25) returns True if either x > 10 or y < 25 is true.
not
Inverts the truth value of a condition. For example, not (x > 10) returns True if x > 10 is false.
